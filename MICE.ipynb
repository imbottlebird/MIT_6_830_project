## Multivariate imputation via chained equations (mice)

Minimal experiment with MICE on iris data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
```

```python
iris = pd.read_csv("datasets/iris_data.csv")
```

```python
iris['class'] = LabelEncoder().fit_transform(iris['class'])
```

```python
iris.head(20)
```

```python
iris.loc[iris.sample(frac=0.1, random_state = 418).index, "petal length"] = np.nan
iris.loc[iris.sample(frac=0.1, random_state = 418).index, "sepal width"] = np.nan
```

```python
iris_MV = iris
```

```python
iris = pd.read_csv("datasets/iris_data.csv")
iris['class'] = LabelEncoder().fit_transform(iris['class'])
```

```python
idx = iris_MV.index[iris_MV['petal length'].isna() | iris_MV['sepal width'].isna()]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp_mean = IterativeImputer(estimator=RandomForestRegressor(), random_state=0)
imp_mean.fit(iris_MV)
display(pd.DataFrame(imp_mean.transform(iris_MV)))
imputed = pd.DataFrame(imp_mean.transform(iris_MV))
```

```python
y_test = iris.iloc[idx][['sepal width','petal length']]
```

```python
idx
```

```python
predicted = imputed.iloc[idx][[1,2]]
```

```python
predicted
```

```python
y_test
```

```python
r2_score(predicted, y_test)
```
