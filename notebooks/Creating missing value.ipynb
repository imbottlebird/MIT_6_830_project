### Creating missing values at random

```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("iris_data.csv")
```

```python
iris.head(30)
```

I am arbitrarily choosing "petal length" column to create 15 percents of missing values on, using a fixed random state of <b>418

```python
iris.loc[iris.sample(frac=0.15, random_state = 418).index, "petal length"] = np.nan
```

```python
iris.head(30)
```

Export the new dataset with missing values

```python
compression_opts = dict(method='zip',
                        archive_name='iris_with_MV.csv')  
iris.to_csv('out.zip', index=False,
          compression=compression_opts)  
```
